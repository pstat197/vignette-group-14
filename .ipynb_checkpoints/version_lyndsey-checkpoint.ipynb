# From Here Down deals with downloading the corpus

Next we will install a module that makes it easy to use manipulate data contained within the zipfile.

This source gives instructions on how to use the "Gutenberg Dammit" tool: https://github.com/aparrish/gutenberg-dammit/

The following chunk grabs the text from one text pulled from the archive

```python
from gutenbergdammit.ziputils import retrieve_one
text = retrieve_one("data/gutenberg-poetry-v001.ndjson.zip", "123/12345.txt")
text[:200]
```

To retrieve the metadata file:

```python
from gutenbergdammit.ziputils import loadmetadata
metadata = loadmetadata("data/gutenberg-poetry-v001.ndjson.zip")
metadata[469]['Title']
```

To search for and retrieve files whose metadata contains particular strings:

```python
from gutenbergdammit.ziputils import searchandretrieve
for info, text in searchandretrieve("data/gutenberg-poetry-v001.ndjson.zip", {'Author': 'Walt Whitman', 'Language':'English'}):
    print(info['Title'][0], "-  gid: ", info['Num'], "  -  length in characters: ", len(text))
    print(text[:1000])
```

```python
from gutenbergdammit.ziputils import searchandretrieve
for info, text in searchandretrieve("data/gutenberg-poetry-v001.ndjson.zip", {'Subject': 'Poetry', 'Language':'English'}):
    print(info['Title'][0], "-  gid: ", info['Num'], "  -  length in characters: ", len(text))
```

# Markov Training on poems with specific subject

```python
len(pristine_lines)
```

```python
love_poem = "\n".join([line for line in random.sample(pristine_lines, 713)])
```

```python
love_model = markovify.NewlineText(love_poem)
```

```python
for i in range(14):
    print(love_model.make_sentence())
```
